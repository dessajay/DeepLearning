```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
#y = ax + b
def linear_function(x):
    a = 0.5 
    b = 2
    
    return a * x + b
```

```python
print(linear_function(5))
```

```python
x = np.arange(-5,5,0.1)
y = linear_function(x)
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
plt.show()
```

```python
#y = ax2 + bx + c
def quadratic_function(x):
    a = 1
    b = -1
    c = -2
    return a*x**2 + b*x + c 
```

```python
print(quadratic_function(2))
```

```python
x = np.arange(-5,5,0.1)
y = quadratic_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic_Function')
plt.show()
```

```python
# y = ax3 + bx2 + cx + d
def cubic_function(x):
    a = 4 
    b = 0 
    c = -1
    d = -8
    
    return a*x**3 + b*x**2 + c*x + d 
```

```python
print(cubic_function(3))
```

```python
x = np.arange(-5,5,0.1)
y = cubic_function(x)

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Cubic_Function")
plt.show()
```

```python
def my_func(x):
    a = 1
    b = -3
    c = 10
    
    return a*x**2 + b*x + c 
```

```python
x = np.arange(-10,10,0.1)
y = my_func(x)

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("My_Function")
plt.scatter(1.5, my_func(1.5))
plt.text(0, my_func(1.5)+10, 'min value of f(x)\n{}'.format(1.5, my_func(1.5)))
plt.show()
```

```python
def exponential_function(x):
    a = 4
    
    return a**x
```

```python
print(exponential_function(4))
print(exponential_function(0))
```

```python
x = np.arange(-3,2,0.1)
y = exponential_function(x)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-4,3)
plt.ylim(-1,15)
plt.title("Exponential_Function")
plt.show()
```

```python
def numerical_defferntial(f,x):
    h = 1e-4
    return (f(x+h)-f(x-h)) / (2*h)
```

```python
def my_func(x):
    return 2*x**2 + 3*x + 2
```

```python
def linear_func(a,b,c,x):
    return c*(x-a) + b
```

```python
c = numerical_defferntial(my_func, 1)
```

```python
x = np.arange(-5,5,0.01)
y = linear_func(1, my_func(1),c,x)

plt.xlim(-5,5)
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, my_func(x), x, y, 'r--')
plt.title('f(x) & linear function')
plt.show()
```
